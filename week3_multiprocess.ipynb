## Python Multiprocess
파이썬은 하나의 프로세스에서 실질적으로 하나의 쓰레드 밖에 사용할 수 없습니다.  
바로 global interpreter lock이라는 장치 때문인데요,  
이 경우 여러 개의 CPU 코어를 활용하지 못하는 한계점이 있습니다.

때문에 파이썬에서 병렬 처리를 하고 싶을 때는 멀티 쓰레드 대신 멀티 프로세스 방식을 사용하게 됩니다.  
오늘은 파이썬에서 멀티 프로세스를 이용한 병렬 처리를 어떻게 구현할 수 있는지 알아보겠습니다.

먼저 이해를 돕기 위해서 싱글 프로세스를 사용하여 숫자를 집계하는 프로그램을 작성해보겠습니다.

```python
import time

start = time.time()
def count(name):
    for i in range(500):
        print(name, " : ", i + 1)

names = ['p1', 'p2', 'p3', 'p4']
for name in names:
    count(name)

end = time.time()
print('timecost : ', end - start)
```

p1부터 p4까지 순차적으로 1부터 50000까지 집계하는 것이 보이시나요?  
그렇다면 이제 멀티 프로세스의 Pool을 이용해서 p1, p2, p3, p4가 동시에 진행되도록 해보겠습니다.  

```python
import multiprocessing
import time

start = time.time()

def count(name):
    for i in range(500):
        print(name, ' : ', i)

names = ['p1', 'p2', 'p3', 'p4']
pool = multiprocessing.Pool(processes=4)
pool.map(count, names)
pool.close()
pool.join()

end = time.time()
print('timecost : ', end - start)
```

마지막으로 multiprocess의 Process 모듈을 사용하여서 멀티 프로세싱을 구현해보겠습니다.

```python
from multiprocessing import Process
import time
import os

start = time.time()

def count():
    pid = os.getpid()
    for i in range(500):
        print('pid :', pid, ' : ', i)

procs = []
process_num = 4
for i in range(process_num):
    proc = Process(target=count)
    procs.append(proc)
    proc.start()

for proc in procs:
    proc.join()

end = time.time()
print('timecost : ', end - start)
```
